```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d=pd.read_csv("C:\\Users\\Manu\\Downloads\\archive (13).zip")
df=pd.DataFrame(d)
df
```

```python
df['PatientId']=df['PatientId'].astype('int64')
```

```python
df
```

```python
df.head(50)
```

```python
df['Gender']=df['Gender'].replace({'F':'Female','M':'Male'})
```

```python
df
```

```python
columns_to_replace=['Scholarship','Hipertension','Diabetes','Alcoholism','Handcap','SMS_received']
df[columns_to_replace]=df[columns_to_replace].replace({1:'Yes',0:'No'})
df.head(50)
```

```python
df
```

```python
df['ScheduledDay']=pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay']=pd.to_datetime(df['AppointmentDay'])
```

```python
df['ScheduledDay']=df['ScheduledDay'].dt.strftime('%d-%m-%y')
df['AppointmentDay']=df['AppointmentDay'].dt.strftime('%d-%m-%y')
```

```python
df
```

```python
df.head(50)
```

## Bar Plot

```python
sns.countplot(data=df,x='Gender')
plt.title("Number of Patients by Gender")
plt.show()
```

## Histogram

```python
plt.hist(df['Age'],bins=20,color='skyblue',edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## Box Plot

```python
sns.boxplot(data=df,x='Gender',y="Age")
plt.title("Age Distribution by Gender")
plt.show()
```

## Pie Chart

```python
sms_counts=df['SMS_received'].value_counts()
plt.pie(sms_counts,labels=sms_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('SMS Received')
plt.axis('equal')
plt.show()
        
```

## Heatmap

```python
df_corr=df.replace({'Yes':1,'No':0})
corr=df_corr.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
